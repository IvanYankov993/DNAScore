```python
import os
import tarfile
```

```python
def create_directory_tree(resample,model):
    # Create the directory tree
    path=f"dir1/resamepl_{resample}/model_{model}"
    os.makedirs(path, exist_ok=True)

    # Create sample files
    with open(f"{path}/grid.csv", "w") as f:
        f.write(f"Content of grid.csv")
    with open(f"{path}/true_pred.csv", "w") as f:
        f.write(f"Content of true_pred.csv")
    with open(f"{path}/x.csv", "w") as f:
        f.write(f"Content of x.csv")

def create_tar_archive():
    # Create a tar.gz archive
    with tarfile.open("example_archive.tar", "w") as archive:
        archive.add("dir1", arcname=os.path.basename("dir1"))
```

```python
def append_to_tar_archive():
    # Append new content to the existing tar.gz archive
    with tarfile.open("example_archive.tar", "a") as archive:
        # Create a new resample directory with model_1 and its CSV files
        new_content_path = "dir1/resample/model_1"
        os.makedirs(new_content_path, exist_ok=True)
        for i in range(1, 4):
            with open(f"{new_content_path}/grid_{i}.csv", "w") as f:
                f.write(f"Appended Content of grid_{i}.csv")
            with open(f"{new_content_path}/true_pred_{i}.csv", "w") as f:
                f.write(f"Appended Content of true_pred_{i}.csv")
            with open(f"{new_content_path}/x_{i}.csv", "w") as f:
                f.write(f"Appended Content of x_{i}.csv")

        # Add the new content to the existing archive
        archive.add(new_content_path, arcname=os.path.basename(new_content_path))
        
def display_tar_contents(file_path):
    with tarfile.open(file_path, "r") as archive:
        print("Contents of the tar file:")
        archive.list()
        # archive.getnames()
        # for member in archive.getmembers():
        #     print(member.name)

def delete_file_from_archive(file_to_delete):
    # Create a new tar.gz archive excluding the specified file
    with tarfile.open("example_archive_updated.tar", "w") as new_archive:
        with tarfile.open("example_archive.tar", "r") as old_archive:
            for member in old_archive.getmembers():
                if member.name != file_to_delete:
                    # Add all files except the one to delete
                    new_archive.addfile(member, old_archive.extractfile(member))
```

```python
for resample in range(1,3):
    for model in range(1,2):
        # Create the directory tree
        create_directory_tree(resample,model)

# Create the tar.gz archive
create_tar_archive()
```

```python
# Append new content to the existing archive
append_to_tar_archive()
```

```python

delete_file_from_archive('model_1/*')
# Display the contents of the appended tar.gz file
display_tar_contents("example_archive_updated.tar")
```
